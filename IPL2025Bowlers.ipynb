```python
import pandas as pd
df = pd.read_csv("IPL2025Bowlers.csv")
print("Preview:\n", df.head())

print("\nShape:", df.shape)
print("\nInfo:")
print(df.info())
print("\nSummary Statistics:\n", df.describe(include="all"))
print("\nMissing Values:\n", df.isnull().sum())
print("\nDuplicate Records:", df.duplicated().sum())
for col in df.select_dtypes(include="object"):
    print(f"\nValue counts for {col}:\n", df[col].value_counts())
    print(df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("IPL2025Bowlers.csv")
df.hist(figsize=(12, 8), bins=20)
plt.suptitle("Histograms of Numerical Features")
plt.show()
plt.figure(figsize=(12, 6))
sns.boxplot(data=df.select_dtypes(include="number"))
plt.title("Boxplots of Numerical Features")
plt.xticks(rotation=45)
plt.show()
num_cols = df.select_dtypes(include="number").columns[:4]
sns.pairplot(df[num_cols], diag_kind="hist")
plt.suptitle("Scatter Plots of Numerical Features", y=1.02)
plt.show()
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df[num_cols[0]], y=df[num_cols[1]], hue=df[num_cols[2]])
plt.title(f"Scatter Plot: {num_cols[0]} vs {num_cols[1]}")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("IPL2025Bowlers.csv")
corr = df.corr(numeric_only=True)

print("Correlation Matrix:\n", corr)

plt.figure(figsize=(10, 6))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", cbar=True)
plt.title("Correlation Heatmap")
plt.show()
num_cols = df.select_dtypes(include="number").columns[:4]
sns.pairplot(df[num_cols], diag_kind="kde")
plt.suptitle("Pairwise Patterns", y=1.02)
plt.show()
for col in df.select_dtypes(include="object"):
    plt.figure(figsize=(10, 5))
    sns.boxplot(x=col, y=df.select_dtypes(include="number").columns[0], data=df)
    plt.title(f"{df.select_dtypes(include='number').columns[0]} vs {col}")
    plt.xticks(rotation=45)
    plt.show()
    print(df)
```
